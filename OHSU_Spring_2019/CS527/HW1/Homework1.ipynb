The IEX Group is a New York based market exchange that is the first founded on the princaple of ensureing fair participation in the market place.  It pervents high frequency trading firms from front running orders by slowing down the market just enough to allow all customers to have access to the same data at the same time. While other exchanges and firms utilize their exchanges and dark pools to pit high frequency traders and traditional traders against eachother to build profits, IEX attempts to win market share by offering a more fair and transparent market and a level playing feild.
