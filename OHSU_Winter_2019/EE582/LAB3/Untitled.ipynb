```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
from scipy.signal import lfilter, cheby1
from scipy.fftpack import fft, ifft
from scipy.signal import convolve
```

### Application of FIR filter to artificial signal

In this section an attempt was made to seperate the signal appart using a low pass filter. I was not able to do this because the FIR method is not sensitive enough. 

```python
# load artifical siganl into x
artifical = loadmat('br_hr.mat')
x = artifical['br_hr']
x = x.squeeze()
plt.plot(x)
x.shape
```

```python
# build FIR filter from hanning window
b = np.hanning(30)
a = np.array([1])

# apply filter to input signal
y =lfilter(x=x,a=a,b=b)

# view filtered signal
plt.plot(y)
```

```python

# view fft of signal to determine rates
plt.plot(abs(fft(y))[1:101])
```

### Application of IRR Filter to Artifical Signal

```python
b, a = cheby1(7, .2 , .004, 'low')


# apply filter to input signal
y =lfilter(x=x,a=a,b=b)

# view filtered signal
plt.plot(y[:])
#print(y.shape)
```

```python

# view fft of signal to determine rates
plt.plot(abs(fft(y))[:10])
```
