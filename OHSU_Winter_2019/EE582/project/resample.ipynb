```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample, resample_poly
```

```python
f = 40
x = np.arange(100)
sig = np.sin(2*np.pi*f*x/1000)
```

```python
plt.plot(sig)
```

```python
re_sig = resample(sig,115)
plt.plot(re_sig)
```

```python
re_poly_sig = resample_poly(x,105,100)
plt.plot(re_poly_sig)
```
