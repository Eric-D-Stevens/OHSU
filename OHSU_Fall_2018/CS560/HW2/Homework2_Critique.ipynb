$ KB = \{p(X)\} $

The ground instance we obtain from the bottom-up proof procedure is:

yes ← $p(X)$

Therefore it can be said that <b> $KB$ implies $p(a)$ </b>

Furthermore we can say that because we must subsitute constants for variables uniformly, any constant that is to exist in our syntax will have a similar resulting ground instance. The introduction of a statment like $p(X)$ into a knowledge base will double the number of interpretations but do nothing to the number of models.
