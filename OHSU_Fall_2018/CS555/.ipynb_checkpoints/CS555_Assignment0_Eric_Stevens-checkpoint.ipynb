### Use Pynini

### Using Matplotlib and Numpy
code taken form matplotlib ("https://matplotlib.org/2.0.2/examples/lines_bars_and_markers/barh_demo.html")

```python


import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

The code above leverages the **Matplotlib** python package to generate a visualization for data. The data itself is randomly generated by use of the **Numpy** library, which contains a random number generator.
