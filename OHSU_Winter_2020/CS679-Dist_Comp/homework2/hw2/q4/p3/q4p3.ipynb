```python
import pandas as pd
import pickle
```

```python
with open('oregon_top.pkl', 'rb') as f:
    d = pickle.load(f)

l = []
for i in d:
       l.append((i[1], i[0].split('\t')[1]))
    
t = pd.DataFrame(l)
t.columns = ['Portland Oregon Downloads', 'Article']
t
```
